```python
import re
import requests
from rdkit import Chem
from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')

# -----------------------
# Lipid notation handler
# -----------------------
def expand_lipid_notation(name):
    """
    Convert shorthand lipid notation to searchable names.
    Returns None if the lipid is too generic/complex for SMILES lookup.
    """
    # Carnitine derivatives
    if 'carnitine' in name.lower():
        m = re.match(r'C(\d+):(\d+)\s+carnitine', name, re.I)
        if m:
            carbons, double_bonds = m.groups()
            # Convert to acylcarnitine names
            acyl_names = {
                ('2', '0'): 'acetylcarnitine',
                ('4', '0'): 'butyrylcarnitine', 
                ('14', '0'): 'tetradecanoylcarnitine',
                ('16', '0'): 'palmitoylcarnitine',
                ('18', '0'): 'stearoylcarnitine',
                ('18', '1'): 'oleoylcarnitine',
                ('18', '2'): 'linoleoylcarnitine',
                ('20', '4'): 'arachidonoylcarnitine',
            }
            return acyl_names.get((carbons, double_bonds))
        if re.match(r'C(\d+)\s+carnitine', name, re.I):
            m = re.match(r'C(\d+)\s+carnitine', name, re.I)
            carbons = m.group(1)
            names = {'2': 'acetylcarnitine', '4': 'butyrylcarnitine', '14': 'tetradecanoylcarnitine',
                     '16': 'palmitoylcarnitine', '18': 'stearoylcarnitine', '20': 'eicosanoylcarnitine'}
            return names.get(carbons)
        # Special carnitine derivatives
        if 'C3-DC-CH3' in name:
            return '3-methylglutarylcarnitine'
    
    # Cholesteryl esters - convert to specific fatty acid esters
    if 'CE' in name and re.match(r'C(\d+):(\d+)\s+CE', name):
        m = re.match(r'C(\d+):(\d+)\s+CE', name)
        carbons, double_bonds = m.groups()
        fatty_acids = {
            ('14', '0'): 'cholesteryl myristate',
            ('16', '0'): 'cholesteryl palmitate',
            ('16', '1'): 'cholesteryl palmitoleate',
            ('18', '0'): 'cholesteryl stearate',
            ('18', '1'): 'cholesteryl oleate',
            ('18', '2'): 'cholesteryl linoleate',
            ('18', '3'): 'cholesteryl linolenate',
            ('20', '3'): 'cholesteryl eicosatrienoate',
            ('20', '4'): 'cholesteryl arachidonate',
            ('20', '5'): 'cholesteryl eicosapentaenoate',
            ('22', '4'): 'cholesteryl adrenate',
            ('22', '5'): 'cholesteryl docosapentaenoate',
            ('22', '6'): 'cholesteryl docosahexaenoate',
        }
        return fatty_acids.get((carbons, double_bonds))
    
    # Most other complex lipids (SM, LPC, DAG, TAG, PC, PE, etc.) are too complex
    # and generic for simple SMILES lookup - would need specific structures
    if any(x in name for x in ['SM', 'LPC', 'LPE', 'MAG', 'DAG', 'TAG', 'PC', 'PE', 'plasmalogen', 'ceramide']):
        return None  # Skip - too complex/generic
    
    return name

# -----------------------
# Name preprocessing
# -----------------------
def preprocess_name(name):
    """Clean up name before SMILES lookup."""
    if not name or not isinstance(name, str):
        return name
    
    # Handle "/" separated names - take first part
    if '/' in name:
        name = name.split('/')[0].strip()
    
    # Expand lipid notation
    expanded = expand_lipid_notation(name)
    if expanded is None:
        return None  # Signal to skip this compound
    if expanded != name:
        name = expanded
    
    # Common name fixes
    name_fixes = {
        'acetytyrosine': 'N-acetyltyrosine',
        'glucurote': 'glucuronic acid',
        'succite': 'succinate',
        'pantothete': 'pantothenate',
        'undecanedionate': 'undecanedioate',
        'methylbutyric aid': 'methylbutyric acid',
        'ADMA': 'NG,NG-dimethyl-L-arginine',
        'SDMA': 'NG,N\'G-dimethyl-L-arginine', 
        'NMMA': 'N-monomethyl-L-arginine',
        'threosphingosine': 'sphinganine',
        'alpha-muricholate': 'alpha-muricholic acid',
        'hyodeoxycholate': 'hyodeoxycholic acid',
        'ursodeoxycholate': 'ursodeoxycholic acid',
        'ketodeoxycholate': '7-ketodeoxycholic acid',
        '7-ketodeoxycholate': '7-ketodeoxycholic acid',
        'trimethylbenzene': '1,2,3-trimethylbenzene',
        'leucine': 'L-leucine',
        'taurochenodesoxycholic acid': 'taurochenodeoxycholic acid',
        'methylpyridonecarboxamide': '1-methyl-2-pyridone-5-carboxamide',
        'L-1,2,3,4-tetrahydro-beta-carboline-3-carboxylic acid': 'tryptophan',
    }
    
    name_lower = name.lower()
    for wrong, right in name_fixes.items():
        if name_lower == wrong.lower():
            return right
    
    return name

# -----------------------
# Simple locant correction
# -----------------------
def fix_numeric_locants(name):
    import re

    # Match: start, then a sequence of digits separated by . or -,
    # then a final hyphen before the word.
    m = re.match(r"^(\d+(?:[.-]\d+)*)(-)([A-Za-z].*)$", name)
    if not m:
        return name

    locants, final_hyphen, rest = m.groups()

    # Convert all inner separators to commas
    locants = re.sub(r"[.-]", ",", locants)

    return locants + final_hyphen + rest

# -----------------------
# Simple anion→acid mapper
# -----------------------
ACID_MAP = {
    "urate": "uric acid",
    "lactate": "lactic acid",
    "citrate": "citric acid",
    "succinate": "succinic acid",
    "fumarate": "fumaric acid",
    "malate": "malic acid",
    "oxalate": "oxalic acid",
    "tartrate": "tartaric acid",
    "pyruvate": "pyruvic acid",
    "benzoate": "benzoic acid",
    "acetate": "acetic acid",
    "propionate": "propionic acid",
    "butyrate": "butyric acid",
    "valerate": "valeric acid",
    "caproate": "caproic acid",
    "cholate": "cholic acid",
    "chenodeoxycholate": "chenodeoxycholic acid",
    "deoxycholate": "deoxycholic acid",
    "eicosapentaenoate": "eicosapentaenoic acid",
    "docosapentaenoate": "docosapentaenoic acid",
    "eicosatrienoate": "eicosatrienoic acid",
    "eicosenoate": "eicosenoic acid",
}

def convert_anion(name):
    for k,v in ACID_MAP.items():
        if name.lower().endswith(k):
            return name[: -len(k)] + v
    return name

# -----------------------
# SMILES validation
# -----------------------
def is_valid_smiles(s):
    return bool(s) and Chem.MolFromSmiles(s) is not None

# -----------------------
# Resolvers
# -----------------------
def chembl_name_to_smiles(name):
    try:
        url = f"https://www.ebi.ac.uk/chembl/api/data/molecule?pref_name={name}"
        r = requests.get(url, timeout=6)
        if r.status_code != 200:
            return None
        data = r.json().get("molecules", [])
        if not data:
            return None
        return data[0].get("molecule_structures", {}).get("canonical_smiles")
    except:
        return None

def pubchem_name_to_smiles(name):
    try:
        url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{name}/property/IsomericSMILES/JSON"
        r = requests.get(url, timeout=6)
        if r.status_code != 200:
            return None
        return r.json()["PropertyTable"]["Properties"][0]["IsomericSMILES"]
    except:
        return None

def cactus_name_to_smiles(name):
    try:
        url = f"https://cactus.nci.nih.gov/chemical/structure/{name}/smiles"
        r = requests.get(url, timeout=6)
        if r.status_code != 200:
            return None
        t = r.text.strip()
        return None if "Not Found" in t else t
    except:
        return None

RESOLVERS = (
    chembl_name_to_smiles,
    pubchem_name_to_smiles,
    cactus_name_to_smiles,
)

# -----------------------
# Master function
# -----------------------
def name_to_smiles(name):
    # Preprocess the name
    cleaned = preprocess_name(name)
    if cleaned is None:
        # Skip complex/generic lipids
        return None
    
    orig = cleaned
    fixed = convert_anion(fix_numeric_locants(cleaned))

    for candidate in (orig, fixed):
        for resolver in RESOLVERS:
            s = resolver(candidate)
            if is_valid_smiles(s):
                return s
    print(f"No SMILES found for {name}")
    return None

SMILES = [name_to_smiles(x) for x in conversion['Exact Match to Standard (* = isomer family)'].to_list()]
conversion['SMILES'] = SMILES
conversion.to_csv(data_dir+'/metabolites_to_SMILES.csv')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
from scipy.cluster.hierarchy import dendrogram, linkage

# Try to import UMAP, handle version issues
try:
    import umap
    UMAP_AVAILABLE = True
except:
    UMAP_AVAILABLE = False

# ======================
# 1. DATA PREPARATION
# ======================

# Load data
embeddings = pd.read_csv('/home/ubuntu/ibd_data/mole_embeddings.csv', index_col=0)
conversion = pd.read_csv('/home/ubuntu/ibd_data/metabolites_to_SMILES.csv')

# Filter conversion for valid SMILES (not None/NaN)
conversion_valid = conversion[conversion['SMILES'].notna()].copy()

# Keep only embeddings that match valid SMILES
valid_smiles = set(conversion_valid['SMILES']) & set(embeddings.index)
embeddings_clean = embeddings.loc[list(valid_smiles)]
conversion_valid = conversion_valid[conversion_valid['SMILES'].isin(valid_smiles)]

# Create mapping: SMILES -> metabolite name
smiles_to_name = dict(zip(conversion_valid['SMILES'], 
                          conversion_valid['Exact Match to Standard (* = isomer family)']))

print(f"Total metabolites with embeddings: {len(embeddings_clean)}")

# ======================
# 2. CLASSIFY METABOLITES BY CHEMICAL CLASS
# ======================

def classify_metabolite(name):
    """Classify metabolite by chemical class based on name."""
    name_lower = str(name).lower()
    
    # Amino acids
    amino_acids = ['alanine', 'arginine', 'asparagine', 'aspartic', 'aspartate', 
                   'cysteine', 'cystine', 'glutamic', 'glutamate', 'glutamine',
                   'glycine', 'histidine', 'isoleucine', 'leucine', 'lysine',
                   'methionine', 'phenylalanine', 'proline', 'serine', 'threonine',
                   'tryptophan', 'tyrosine', 'valine', 'carnosine', 'anserine',
                   'citrulline', 'ornithine', 'dimethylarginine', 'adma', 'sdma']
    
    # Bile acids
    bile_acids = ['cholate', 'chenodeoxycholate', 'deoxycholate', 'lithocholate',
                  'ursodeoxycholate', 'hyodeoxycholate', 'muricholate',
                  'taurocholate', 'taurodeoxycholate', 'taurochenodeoxycholate',
                  'glycocholate', 'ketodeoxycholate', 'tauroursodeoxycholate']
    
    # Fatty acids
    fatty_acids = ['palmitic', 'stearic', 'oleic', 'linoleic', 'linolenic',
                   'arachidonic', 'docosahexaenoic', 'eicosapentaenoic',
                   'myristic', 'lauric', 'capric', 'caprylic', 'caproic',
                   'butyric', 'butyrate', 'valeric', 'propionate', 'acetate',
                   'hexanoic', 'octanoic', 'decanoic', 'eicosatrienoic',
                   'docosapentaenoic', 'eicosenoic', 'palmitoleic', 'adrenate',
                   'dodecanedioic', 'undecanedioate']
    
    # Carnitines
    carnitines = ['carnitine', 'acylcarnitine', 'butyrylcarnitine', 
                  'palmitoylcarnitine', 'oleoylcarnitine', 'tetradecanoyl',
                  'glutarylcarnitine']
    
    # Nucleotides/bases
    nucleotides = ['adenine', 'guanine', 'cytosine', 'thymine', 'uracil',
                   'adenosine', 'guanosine', 'cytidine', 'uridine', 'thymidine',
                   'inosine', 'hypoxanthine', 'xanthine', 'urate', 'deoxyinosine']
    
    # Carbohydrates/sugars
    carbohydrates = ['glucose', 'fructose', 'galactose', 'mannose', 'ribose',
                     'xylose', 'arabinose', 'lactose', 'maltose', 'sucrose',
                     'trehalose', 'fucose', 'glucuronic', 'levoglucosan']
    
    # Organic acids (TCA cycle, etc)
    organic_acids = ['citric', 'citrate', 'succinic', 'succinate', 'fumaric', 'fumarate',
                     'malic', 'malate', 'oxaloacetic', 'pyruvic', 'pyruvate',
                     'lactic', 'lactate', 'ketoglutaric', 'oxalic', 'tartaric',
                     'erythronic', 'glyceric', 'threonic']
    
    # Sterols/steroids
    sterols = ['cholesterol', 'cholesteryl', 'cholestenone', 'ergosterol',
               'stigmasterol', 'sitosterol', 'coprostanol', 'epoxysqualene']
    
    # Complex lipids
    complex_lipids = ['phosphatidyl', 'lysophosphatidyl', 'sphingomyelin',
                      'ceramide', 'glycerol', 'diacylglycerol', 'triacylglycerol',
                      'monoacylglycerol', 'sphingosine', 'threosphingosine']
    
    # Vitamins
    vitamins = ['retinol', 'tocopherol', 'niacin', 'riboflavin', 'thiamine',
                'pyridoxine', 'biotin', 'folate', 'pantothenic', 'cobalamin']
    
    # Polyamines
    polyamines = ['putrescine', 'spermidine', 'spermine', 'cadaverine', 'diacetylspermine']
    
    # Phenolics/aromatics
    phenolics = ['benzoic', 'hippuric', 'phenyl', 'indole', 'skatole',
                 'cresol', 'phenol', 'catechol', 'quinone', 'hydroxybenzoic',
                 'dihydroxybenzoic', 'urocanic', 'urobilin']
    
    # Choline compounds
    cholines = ['choline', 'betaine', 'trimethylamine', 'dmg']
    
    # Check each category (order matters - check specific before general)
    if any(ca in name_lower for ca in carnitines):
        return 'Acylcarnitines'
    elif any(ba in name_lower for ba in bile_acids):
        return 'Bile Acids'
    elif any(aa in name_lower for aa in amino_acids):
        return 'Amino Acids'
    elif any(fa in name_lower for fa in fatty_acids):
        return 'Fatty Acids'
    elif any(nt in name_lower for nt in nucleotides):
        return 'Nucleotides/Bases'
    elif any(ch in name_lower for ch in carbohydrates):
        return 'Carbohydrates'
    elif any(oa in name_lower for oa in organic_acids):
        return 'Organic Acids'
    elif any(st in name_lower for st in sterols):
        return 'Sterols'
    elif any(cl in name_lower for cl in complex_lipids):
        return 'Complex Lipids'
    elif any(vt in name_lower for vt in vitamins):
        return 'Vitamins'
    elif any(pa in name_lower for pa in polyamines):
        return 'Polyamines'
    elif any(ph in name_lower for ph in phenolics):
        return 'Phenolics/Aromatics'
    elif any(cho in name_lower for cho in cholines):
        return 'Choline Compounds'
    else:
        return 'Other'

# Classify all metabolites
metabolite_classes = []
for smiles in embeddings_clean.index:
    name = smiles_to_name.get(smiles, '')
    metabolite_classes.append(classify_metabolite(name))

embeddings_clean['Class'] = metabolite_classes

# Print class distribution
print("\nMetabolite class distribution:")
print(embeddings_clean['Class'].value_counts())

# Create color palette
classes = embeddings_clean['Class'].unique()
n_classes = len(classes)
palette = sns.color_palette('husl', n_classes)
class_colors = dict(zip(classes, palette))

# ======================
# 3. DIMENSIONALITY REDUCTION
# ======================

X = embeddings_clean.drop('Class', axis=1).values

# PCA
pca = PCA(n_components=min(50, X.shape[1], X.shape[0]))
pca_result = pca.fit_transform(X)
pca_2d = PCA(n_components=2)
pca_2d_result = pca_2d.fit_transform(X)

# UMAP (with error handling)
if UMAP_AVAILABLE:
    try:
        reducer = umap.UMAP(n_neighbors=min(15, X.shape[0]-1), min_dist=0.1, random_state=42)
        umap_result = reducer.fit_transform(X)
        UMAP_SUCCESS = True
    except Exception as e:
        print(f"UMAP failed: {e}")
        print("Skipping UMAP visualizations...")
        UMAP_SUCCESS = False
else:
    print("UMAP not available, skipping UMAP visualizations...")
    UMAP_SUCCESS = False

# t-SNE
perplexity = min(30, (X.shape[0] - 1) // 3)
tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity)
tsne_result = tsne.fit_transform(pca_result)

# K-means clustering
n_clusters = min(10, X.shape[0] // 5)  # At least 5 samples per cluster
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)
embeddings_clean['Cluster'] = cluster_labels

# ======================
# 4. SEPARATE PLOTS
# ======================

# Plot 1: PCA Explained Variance
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), linewidth=2)
plt.xlabel('Number of Components', fontsize=12)
plt.ylabel('Cumulative Explained Variance', fontsize=12)
plt.title('PCA Explained Variance', fontsize=14, fontweight='bold')
plt.axhline(y=0.9, color='r', linestyle='--', linewidth=1.5, label='90% variance')
plt.axhline(y=0.8, color='orange', linestyle='--', linewidth=1.5, label='80% variance')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/pca_variance.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot 2: PCA 2D - Colored by Chemical Class
plt.figure(figsize=(10, 8))
for class_name in sorted(classes):
    mask = embeddings_clean['Class'] == class_name
    plt.scatter(pca_2d_result[mask, 0], pca_2d_result[mask, 1], 
               c=[class_colors[class_name]], label=class_name, 
               alpha=0.7, s=50, edgecolors='white', linewidth=0.5)
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%} variance)', fontsize=12)
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%} variance)', fontsize=12)
plt.title('PCA: Metabolites Colored by Chemical Class', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/pca_by_class.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot 3: PCA 2D - Colored by Cluster
plt.figure(figsize=(10, 8))
scatter = plt.scatter(pca_2d_result[:, 0], pca_2d_result[:, 1], 
                     c=cluster_labels, cmap='tab20', 
                     alpha=0.7, s=50, edgecolors='white', linewidth=0.5)
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%} variance)', fontsize=12)
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%} variance)', fontsize=12)
plt.title(f'PCA: K-means Clusters (k={n_clusters})', fontsize=14, fontweight='bold')
plt.colorbar(scatter, label='Cluster')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/pca_by_cluster.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot 4 & 5: UMAP - Only if successful
if UMAP_SUCCESS:
    # Plot 4: UMAP - Colored by Chemical Class
    plt.figure(figsize=(10, 8))
    for class_name in sorted(classes):
        mask = embeddings_clean['Class'] == class_name
        plt.scatter(umap_result[mask, 0], umap_result[mask, 1], 
                   c=[class_colors[class_name]], label=class_name, 
                   alpha=0.7, s=50, edgecolors='white', linewidth=0.5)
    plt.xlabel('UMAP 1', fontsize=12)
    plt.ylabel('UMAP 2', fontsize=12)
    plt.title('UMAP: Metabolites Colored by Chemical Class', fontsize=14, fontweight='bold')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig('/home/ubuntu/ibd_data/umap_by_class.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Plot 5: UMAP - Colored by Cluster
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(umap_result[:, 0], umap_result[:, 1], 
                         c=cluster_labels, cmap='tab20', 
                         alpha=0.7, s=50, edgecolors='white', linewidth=0.5)
    plt.xlabel('UMAP 1', fontsize=12)
    plt.ylabel('UMAP 2', fontsize=12)
    plt.title(f'UMAP: K-means Clusters (k={n_clusters})', fontsize=14, fontweight='bold')
    plt.colorbar(scatter, label='Cluster')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig('/home/ubuntu/ibd_data/umap_by_cluster.png', dpi=300, bbox_inches='tight')
    plt.show()

# Plot 6: t-SNE - Colored by Chemical Class
plt.figure(figsize=(10, 8))
for class_name in sorted(classes):
    mask = embeddings_clean['Class'] == class_name
    plt.scatter(tsne_result[mask, 0], tsne_result[mask, 1], 
               c=[class_colors[class_name]], label=class_name, 
               alpha=0.7, s=50, edgecolors='white', linewidth=0.5)
plt.xlabel('t-SNE 1', fontsize=12)
plt.ylabel('t-SNE 2', fontsize=12)
plt.title('t-SNE: Metabolites Colored by Chemical Class', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/tsne_by_class.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot 7: t-SNE - Colored by Cluster
plt.figure(figsize=(10, 8))
scatter = plt.scatter(tsne_result[:, 0], tsne_result[:, 1], 
                     c=cluster_labels, cmap='tab20', 
                     alpha=0.7, s=50, edgecolors='white', linewidth=0.5)
plt.xlabel('t-SNE 1', fontsize=12)
plt.ylabel('t-SNE 2', fontsize=12)
plt.title(f't-SNE: K-means Clusters (k={n_clusters})', fontsize=14, fontweight='bold')
plt.colorbar(scatter, label='Cluster')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/tsne_by_cluster.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot 8: Elbow Plot
inertias = []
K_range = range(2, min(21, X.shape[0] // 3))
for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_temp.fit(X)
    inertias.append(kmeans_temp.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia', fontsize=12)
plt.title('Elbow Method for Optimal k', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/elbow_plot.png', dpi=300, bbox_inches='tight')
plt.show()

# ======================
# 5. CLUSTER CHARACTERIZATION
# ======================

print("\n" + "="*80)
print("CLUSTER COMPOSITION BY CHEMICAL CLASS")
print("="*80)

for cluster_id in range(n_clusters):
    cluster_mask = embeddings_clean['Cluster'] == cluster_id
    cluster_classes = embeddings_clean[cluster_mask]['Class'].value_counts()
    total = cluster_mask.sum()
    
    print(f"\n--- Cluster {cluster_id} ({total} metabolites) ---")
    for class_name, count in cluster_classes.items():
        pct = 100 * count / total
        print(f"  {class_name}: {count} ({pct:.1f}%)")
    
    # Show example metabolites
    cluster_smiles = embeddings_clean[cluster_mask].index[:5]
    examples = [smiles_to_name.get(s, s[:30]) for s in cluster_smiles]
    print(f"  Examples: {', '.join(examples)}")

# ======================
# 6. CLASS PURITY OF CLUSTERS
# ======================

# Compute cluster purity (max class percentage in each cluster)
cluster_purities = []
for cluster_id in range(n_clusters):
    cluster_mask = embeddings_clean['Cluster'] == cluster_id
    if cluster_mask.sum() > 0:
        cluster_classes = embeddings_clean[cluster_mask]['Class'].value_counts()
        purity = cluster_classes.max() / cluster_mask.sum()
        cluster_purities.append(purity)
        dominant_class = cluster_classes.idxmax()
    else:
        cluster_purities.append(0)
        dominant_class = 'None'

plt.figure(figsize=(10, 6))
bars = plt.bar(range(n_clusters), cluster_purities, alpha=0.8, edgecolor='black')
plt.xlabel('Cluster ID', fontsize=12)
plt.ylabel('Purity (Dominant Class %)', fontsize=12)
plt.title('Cluster Purity: Fraction of Dominant Chemical Class', fontsize=14, fontweight='bold')
plt.ylim([0, 1])
plt.axhline(y=0.5, color='r', linestyle='--', linewidth=1.5, label='50% threshold')
plt.xticks(range(n_clusters))
plt.legend()
plt.grid(alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/cluster_purity.png', dpi=300, bbox_inches='tight')
plt.show()

# ======================
# 7. CONFUSION MATRIX: Cluster vs Class
# ======================

from sklearn.metrics import confusion_matrix

# Create confusion matrix - convert cluster labels to strings
conf_mat = confusion_matrix(embeddings_clean['Class'], 
                            embeddings_clean['Cluster'].astype(str))

# Get sorted class names and cluster labels as strings
class_names = sorted(embeddings_clean['Class'].unique())
cluster_names = sorted(embeddings_clean['Cluster'].unique())

plt.figure(figsize=(12, 8))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', 
            xticklabels=[f'C{i}' for i in cluster_names],
            yticklabels=class_names)
plt.xlabel('K-means Cluster', fontsize=12)
plt.ylabel('Chemical Class', fontsize=12)
plt.title('Confusion Matrix: Chemical Class vs K-means Cluster', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/class_cluster_confusion.png', dpi=300, bbox_inches='tight')
plt.show()

# Save results
embeddings_clean['Metabolite'] = embeddings_clean.index.map(smiles_to_name)
embeddings_clean[['Metabolite', 'Class', 'Cluster']].to_csv(
    '/home/ubuntu/ibd_data/metabolite_classification.csv'
)

print("\n✓ All plots saved and classification results exported!")
print("✓ Saved to: /home/ubuntu/ibd_data/")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler

print(f"df4 shape: {df4.shape}")
print(f"conversion shape: {conversion.shape}")
print(f"embeddings shape: {embeddings.shape}")

# Extract metabolite data (starting from row 7, zero-indexed)
# Get diagnosis info (appears to be in row 1)
diagnosis = df4.iloc[1]  # Row with "Diagnosis"
metabolite_data = df4.iloc[7:].copy()  # Metabolite abundances start at row 7

print(f"\nDiagnosis counts:")
print(diagnosis.value_counts())
print(f"\nMetabolite data shape: {metabolite_data.shape}")

# Filter conversion for valid SMILES
conversion_valid = conversion[conversion['SMILES'].notna()].copy()
print(f"\nValid SMILES in conversion: {len(conversion_valid)}")

# De-duplicate embeddings by taking first occurrence of each SMILES
embeddings_dedup = embeddings[~embeddings.index.duplicated(keep='first')].copy()
print(f"Embeddings after deduplication: {embeddings_dedup.shape}")

# Create mapping from metabolite index to SMILES
# conversion index contains the metabolite IDs from df3/df4
metabolite_to_smiles = dict(zip(conversion_valid.index, conversion_valid['SMILES']))

# Find metabolites in df4 that have SMILES and embeddings
matched_metabolites = []
matched_smiles = []

for metabolite_id in metabolite_data.index:
    if metabolite_id in metabolite_to_smiles:
        smiles = metabolite_to_smiles[metabolite_id]
        if smiles in embeddings_dedup.index:
            matched_metabolites.append(metabolite_id)
            matched_smiles.append(smiles)

print(f"\nMatched metabolites with both SMILES and embeddings: {len(matched_metabolites)}")

# Extract matched data
metabolite_data_matched = metabolite_data.loc[matched_metabolites]
print(f"Matched metabolite data shape: {metabolite_data_matched.shape}")

# Create sample embeddings using weighted average
samples = metabolite_data_matched.columns
n_samples = len(samples)
embedding_dim = embeddings_dedup.shape[1]

# Pre-compute embeddings for each matched metabolite (handle duplicate SMILES)
# Each metabolite gets its corresponding SMILES embedding
metabolite_embeddings_array = np.zeros((len(matched_metabolites), embedding_dim))
for idx, smiles in enumerate(matched_smiles):
    metabolite_embeddings_array[idx] = embeddings_dedup.loc[smiles].values

print(f"Metabolite embeddings array shape: {metabolite_embeddings_array.shape}")

sample_embeddings = np.zeros((n_samples, embedding_dim))

for i, sample in enumerate(samples):
    # Get abundances for this sample
    abundances = metabolite_data_matched[sample].values.astype(float)
    
    # Replace NaN with 0
    abundances = np.nan_to_num(abundances, nan=0.0)
    
    # Log-normalize: log(x + 1)
    log_abundances = np.log1p(abundances)
    
    # Normalize to sum to 1 (weighted average)
    if log_abundances.sum() > 0:
        weights = log_abundances / log_abundances.sum()
    else:
        weights = np.ones_like(log_abundances) / len(log_abundances)
    
    # Weighted average using pre-computed embeddings
    sample_embeddings[i] = np.dot(weights, metabolite_embeddings_array)

# Create DataFrame with sample embeddings
sample_embeddings_df = pd.DataFrame(sample_embeddings, index=samples)
print(f"\nSample embeddings shape: {sample_embeddings_df.shape}")

# Prepare labels
y_multiclass = diagnosis.values
y_binary = np.where((y_multiclass == 'UC') | (y_multiclass == 'CD'), 'IBD', 'Control')

print(f"\nMulticlass distribution:")
unique, counts = np.unique(y_multiclass, return_counts=True)
for label, count in zip(unique, counts):
    print(f"  {label}: {count}")

print(f"\nBinary distribution:")
unique, counts = np.unique(y_binary, return_counts=True)
for label, count in zip(unique, counts):
    print(f"  {label}: {count}")

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(sample_embeddings_df)

print("\n" + "="*80)
print("MULTICLASS CLASSIFICATION (UC vs CD vs Control)")
print("="*80)

# Split data for multiclass
X_train_mc, X_test_mc, y_train_mc, y_test_mc = train_test_split(
    X_scaled, y_multiclass, test_size=0.2, random_state=42, stratify=y_multiclass
)

# Train multiclass logistic regression
lr_multiclass = LogisticRegression(max_iter=1000, random_state=42, multi_class='multinomial')
lr_multiclass.fit(X_train_mc, y_train_mc)

# Predictions
y_pred_mc = lr_multiclass.predict(X_test_mc)

# Evaluation
print(f"\nTraining set size: {len(X_train_mc)}")
print(f"Test set size: {len(X_test_mc)}")
print(f"\nTest Accuracy: {accuracy_score(y_test_mc, y_pred_mc):.4f}")
print(f"Train Accuracy: {accuracy_score(y_train_mc, lr_multiclass.predict(X_train_mc)):.4f}")

print("\nClassification Report:")
print(classification_report(y_test_mc, y_pred_mc))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_mc, y_pred_mc))

# Cross-validation
cv_scores_mc = cross_val_score(lr_multiclass, X_scaled, y_multiclass, cv=5, scoring='accuracy')
print(f"\n5-Fold CV Accuracy: {cv_scores_mc.mean():.4f} (+/- {cv_scores_mc.std():.4f})")

print("\n" + "="*80)
print("BINARY CLASSIFICATION (IBD vs Control)")
print("="*80)

# Split data for binary
X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(
    X_scaled, y_binary, test_size=0.2, random_state=42, stratify=y_binary
)

# Train binary logistic regression
lr_binary = LogisticRegression(max_iter=1000, random_state=42)
lr_binary.fit(X_train_bin, y_train_bin)

# Predictions
y_pred_bin = lr_binary.predict(X_test_bin)

# Evaluation
print(f"\nTraining set size: {len(X_train_bin)}")
print(f"Test set size: {len(X_test_bin)}")
print(f"\nTest Accuracy: {accuracy_score(y_test_bin, y_pred_bin):.4f}")
print(f"Train Accuracy: {accuracy_score(y_train_bin, lr_binary.predict(X_train_bin)):.4f}")

print("\nClassification Report:")
print(classification_report(y_test_bin, y_pred_bin))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_bin, y_pred_bin))

# Cross-validation
cv_scores_bin = cross_val_score(lr_binary, X_scaled, y_binary, cv=5, scoring='accuracy')
print(f"\n5-Fold CV Accuracy: {cv_scores_bin.mean():.4f} (+/- {cv_scores_bin.std():.4f})")

# Save results
results = {
    'multiclass': {
        'model': lr_multiclass,
        'test_accuracy': accuracy_score(y_test_mc, y_pred_mc),
        'train_accuracy': accuracy_score(y_train_mc, lr_multiclass.predict(X_train_mc)),
        'cv_scores': cv_scores_mc
    },
    'binary': {
        'model': lr_binary,
        'test_accuracy': accuracy_score(y_test_bin, y_pred_bin),
        'train_accuracy': accuracy_score(y_train_bin, lr_binary.predict(X_train_bin)),
        'cv_scores': cv_scores_bin
    }
}

# Save sample embeddings
sample_embeddings_df.to_csv(data_dir + '/sample_embeddings.csv')
print("\n✓ Sample embeddings saved to sample_embeddings.csv")
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, accuracy_score, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

print("="*80)
print("MULTI-CHANNEL APPROACH: Using Embeddings WITHOUT Aggregation")
print("="*80)

# ============================================================================
# PREPARE DATA: Multiple representations
# ============================================================================

# 1. Raw abundances (297 metabolites)
X_abundances = metabolite_data_matched.T.values.astype(float)
X_abundances = np.nan_to_num(X_abundances, nan=0.0)
X_abundances_log = np.log1p(X_abundances)

# 2. Embeddings matrix (297 metabolites × 768 dims)
# For each sample, create a matrix of metabolite embeddings weighted by abundance
def create_embedding_features(abundances, embeddings_array, method='weighted'):
    """Create embedding-based features without full aggregation"""
    n_samples = abundances.shape[0]
    n_metabolites = abundances.shape[1]
    embedding_dim = embeddings_array.shape[1]
    
    if method == 'weighted':
        # Multiply each embedding by its abundance (element-wise)
        features = np.zeros((n_samples, n_metabolites, embedding_dim))
        for i in range(n_samples):
            log_abund = np.log1p(abundances[i])
            features[i] = embeddings_array * log_abund[:, np.newaxis]
        return features.reshape(n_samples, -1)  # Flatten to 2D
    
    elif method == 'pca':
        # Use PCA to reduce embedding dimensions, then weight by abundance
        pca = PCA(n_components=50)
        embeddings_reduced = pca.fit_transform(embeddings_array)
        features = np.zeros((n_samples, n_metabolites, 50))
        for i in range(n_samples):
            log_abund = np.log1p(abundances[i])
            features[i] = embeddings_reduced * log_abund[:, np.newaxis]
        return features.reshape(n_samples, -1), embeddings_reduced, pca
    
    elif method == 'concat':
        # Concatenate abundances with embedding statistics
        features_list = [np.log1p(abundances)]
        for i in range(n_samples):
            log_abund = np.log1p(abundances[i])
            weights = log_abund / (log_abund.sum() + 1e-10)
            weighted_emb = (embeddings_array.T * weights).T
            # Add statistics
            features_list.append([
                weighted_emb.mean(axis=0),  # Mean embedding
                weighted_emb.std(axis=0),   # Std embedding
                weighted_emb.max(axis=0),   # Max embedding
                weighted_emb.min(axis=0)    # Min embedding
            ])
        return np.hstack([arr if arr.ndim == 2 else np.array(arr).reshape(len(arr), -1) 
                         for arr in features_list])

# Create different feature representations
print("\nCreating multi-channel feature representations...")

# Method 1: PCA-reduced embeddings weighted by abundance
X_emb_pca, embeddings_pca, pca_model = create_embedding_features(
    X_abundances, metabolite_embeddings_array, method='pca'
)
print(f"PCA embedding features shape: {X_emb_pca.shape}")

# Method 2: Concatenate abundances with embedding statistics
X_concat = np.hstack([
    X_abundances_log,  # Raw log abundances
    X_emb_pca          # PCA-reduced weighted embeddings
])
print(f"Concatenated features shape: {X_concat.shape}")

# Standardize all feature sets
scaler_pca = StandardScaler()
X_emb_pca_scaled = scaler_pca.fit_transform(X_emb_pca)

scaler_concat = StandardScaler()
X_concat_scaled = scaler_concat.fit_transform(X_concat)

# ============================================================================
# EXTRACT ALL METADATA LABELS
# ============================================================================
print("\n" + "="*80)
print("EXTRACTING PATIENT METADATA FOR MULTIPLE TASKS")
print("="*80)

# Get all metadata rows
metadata_rows = df4.iloc[:7]  # First 7 rows are metadata
print(f"\nMetadata rows: {list(metadata_rows.index)}")

# Extract labels for different tasks
labels = {}

# Diagnosis (multiclass and binary)
labels['diagnosis_multi'] = df4.iloc[1].values  # CD, UC, Control
labels['diagnosis_binary'] = np.where(
    (labels['diagnosis_multi'] == 'UC') | (labels['diagnosis_multi'] == 'CD'), 
    'IBD', 'Control'
)

# Age (continuous) - convert to float
age_raw = df4.iloc[0].values
labels['age'] = np.array([float(x) if pd.notna(x) and str(x).replace('.','').isdigit() 
                          else np.nan for x in age_raw])

# Fecal Calprotectin (continuous)
fecal_raw = df4.iloc[2].values
labels['fecal_calprotectin'] = np.array([float(x) if pd.notna(x) and str(x).replace('.','').replace('e','').replace('-','').replace('+','').isdigit()
                                         else np.nan for x in fecal_raw])

# Binary medication/treatment labels
labels['antibiotic'] = (df4.iloc[3].values == 'Yes').astype(int)
labels['immunosuppressant'] = (df4.iloc[4].values == 'Yes').astype(int)
labels['mesalamine'] = (df4.iloc[5].values == 'Yes').astype(int)
labels['steroids'] = (df4.iloc[6].values == 'Yes').astype(int)

# Print statistics
print("\nLabel statistics:")
for key, val in labels.items():
    if key.endswith('_multi') or key.endswith('_binary'):
        print(f"\n{key}:")
        unique, counts = np.unique(val, return_counts=True)
        for u, c in zip(unique, counts):
            print(f"  {u}: {c}")
    elif key in ['age', 'fecal_calprotectin']:
        valid = ~np.isnan(val)
        print(f"\n{key}:")
        print(f"  Valid samples: {valid.sum()} / {len(val)}")
        print(f"  Range: {val[valid].min():.1f} - {val[valid].max():.1f}")
        print(f"  Mean ± Std: {val[valid].mean():.1f} ± {val[valid].std():.1f}")
    else:
        print(f"\n{key}: Yes={val.sum()}, No={len(val)-val.sum()}")

print("\n" + "="*80)
```

```python
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("TESTING LOGISTIC REGRESSION ON VARIOUS PREDICTION TASKS")
print("="*80)

# ============================================================================
# FUNCTION TO EVALUATE TASKS (Simplified - only logistic/ridge regression)
# ============================================================================

def evaluate_classification_task(X, y, task_name, feature_name):
    """Evaluate classification task with logistic regression"""
    print(f"\nTask: {task_name:30s} | Features: {feature_name}")
    
    try:
        # Handle multiclass vs binary
        if len(np.unique(y)) > 2:
            model = LogisticRegression(max_iter=1000, random_state=42, multi_class='multinomial')
        else:
            model = LogisticRegression(max_iter=1000, random_state=42)
        
        cv_scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
        result = {
            'accuracy': cv_scores.mean(),
            'std': cv_scores.std()
        }
        print(f"  → Accuracy: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
        return {'Logistic Regression': result}
    except Exception as e:
        print(f"  → Error: {str(e)[:60]}")
        return {}

def evaluate_regression_task(X, y, task_name, feature_name):
    """Evaluate regression task with ridge regression"""
    # Remove NaN values
    valid_mask = ~np.isnan(y)
    X_valid = X[valid_mask]
    y_valid = y[valid_mask]
    
    if len(y_valid) < 20:
        print(f"\nTask: {task_name:30s} - SKIPPED (not enough valid samples: {len(y_valid)})")
        return {}
    
    print(f"\nTask: {task_name:30s} | Features: {feature_name} (n={len(y_valid)})")
    
    try:
        model = Ridge(alpha=1.0, random_state=42)
        cv_scores_r2 = cross_val_score(model, X_valid, y_valid, cv=min(5, len(y_valid)//2), scoring='r2')
        cv_scores_mae = cross_val_score(model, X_valid, y_valid, cv=min(5, len(y_valid)//2), 
                                       scoring='neg_mean_absolute_error')
        result = {
            'r2': cv_scores_r2.mean(),
            'r2_std': cv_scores_r2.std(),
            'mae': -cv_scores_mae.mean(),
            'mae_std': cv_scores_mae.std()
        }
        print(f"  → R² = {cv_scores_r2.mean():.4f} ± {cv_scores_r2.std():.4f}, "
              f"MAE = {-cv_scores_mae.mean():.2f} ± {cv_scores_mae.std():.2f}")
        return {'Ridge Regression': result}
    except Exception as e:
        print(f"  → Error: {str(e)[:60]}")
        return {}

# ============================================================================
# RUN EXPERIMENTS: COMPARE FEATURE REPRESENTATIONS
# ============================================================================

all_results = {}

feature_sets = {
    'Raw Abundances (297)': X_matched_scaled,
    'Embeddings (PCA-weighted)': X_emb_pca_scaled,
    'Combined (Abundances + Embeddings)': X_concat_scaled
}

# ============================================================================
# TASK 1: DIAGNOSIS PREDICTION (Binary: IBD vs Control)
# ============================================================================
print("\n" + "#"*80)
print("# TASK 1: IBD vs Control Classification")
print("#"*80)

task_results = {}
for feat_name, X_feat in feature_sets.items():
    task_results[feat_name] = evaluate_classification_task(
        X_feat, labels['diagnosis_binary'], 
        'IBD vs Control', feat_name
    )
all_results['IBD_vs_Control'] = task_results

# ============================================================================
# TASK 2: DIAGNOSIS PREDICTION (Multiclass: UC vs CD vs Control)
# ============================================================================
print("\n" + "#"*80)
print("# TASK 2: UC vs CD vs Control Classification")
print("#"*80)

task_results = {}
for feat_name, X_feat in feature_sets.items():
    task_results[feat_name] = evaluate_classification_task(
        X_feat, labels['diagnosis_multi'], 
        'UC vs CD vs Control', feat_name
    )
all_results['UC_CD_Control'] = task_results

# ============================================================================
# TASK 3: ANTIBIOTIC USAGE PREDICTION
# ============================================================================
print("\n" + "#"*80)
print("# TASK 3: Antibiotic Usage Prediction")
print("#"*80)

task_results = {}
for feat_name, X_feat in feature_sets.items():
    task_results[feat_name] = evaluate_classification_task(
        X_feat, labels['antibiotic'], 
        'Antibiotic Usage', feat_name
    )
all_results['Antibiotic'] = task_results

# ============================================================================
# TASK 4: IMMUNOSUPPRESSANT USAGE PREDICTION
# ============================================================================
print("\n" + "#"*80)
print("# TASK 4: Immunosuppressant Usage Prediction")
print("#"*80)

task_results = {}
for feat_name, X_feat in feature_sets.items():
    task_results[feat_name] = evaluate_classification_task(
        X_feat, labels['immunosuppressant'], 
        'Immunosuppressant Usage', feat_name
    )
all_results['Immunosuppressant'] = task_results

# ============================================================================
# TASK 5: AGE PREDICTION (Regression)
# ============================================================================
print("\n" + "#"*80)
print("# TASK 5: Age Prediction (Regression)")
print("#"*80)

task_results = {}
for feat_name, X_feat in feature_sets.items():
    task_results[feat_name] = evaluate_regression_task(
        X_feat, labels['age'], 
        'Age Prediction', feat_name
    )
all_results['Age'] = task_results

# ============================================================================
# TASK 6: FECAL CALPROTECTIN PREDICTION (Regression)
# ============================================================================
print("\n" + "#"*80)
print("# TASK 6: Fecal Calprotectin Prediction (Regression)")
print("#"*80)

task_results = {}
for feat_name, X_feat in feature_sets.items():
    task_results[feat_name] = evaluate_regression_task(
        X_feat, labels['fecal_calprotectin'], 
        'Fecal Calprotectin', feat_name
    )
all_results['Fecal_Calprotectin'] = task_results

print("\n" + "="*80)
print("EXPERIMENTS COMPLETED")
print("="*80)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import pdist, squareform, cosine
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("NOVEL EMBEDDING-SPECIFIC ANALYSES")
print("="*80)

# ============================================================================
# 1. CHEMICAL SIMILARITY NETWORK ANALYSIS
# ============================================================================
print("\n" + "#"*80)
print("# TASK 1: Metabolite Substitution Patterns")
print("#"*80)

# Build similarity matrix from embeddings
embedding_similarity = cosine_similarity(metabolite_embeddings_array)

# For each metabolite, find its most similar partners
print("\nFinding metabolite pairs with high structural similarity...")

similar_pairs = []
n_metabolites = len(matched_metabolites)

for i in range(n_metabolites):
    for j in range(i+1, n_metabolites):
        if embedding_similarity[i, j] > 0.90:  # High similarity threshold
            similar_pairs.append({
                'metabolite_1': matched_metabolites[i],
                'metabolite_2': matched_metabolites[j],
                'similarity': embedding_similarity[i, j],
                'idx_1': i,
                'idx_2': j
            })

similar_pairs_df = pd.DataFrame(similar_pairs)
print(f"Found {len(similar_pairs)} highly similar metabolite pairs (similarity > 0.90)")

# Analyze substitution patterns in disease
if len(similar_pairs) > 0:
    print("\n" + "="*80)
    print("Metabolic Compensation Analysis")
    print("="*80)
    print("Looking for cases where one metabolite decreases while similar one increases...")
    
    # Separate by diagnosis
    cd_mask = labels['diagnosis_multi'] == 'CD'
    uc_mask = labels['diagnosis_multi'] == 'UC'
    control_mask = labels['diagnosis_multi'] == 'Control'
    
    compensation_patterns = []
    
    for _, pair in similar_pairs_df.head(20).iterrows():  # Check top 20 pairs
        idx1, idx2 = int(pair['idx_1']), int(pair['idx_2'])
        
        # Calculate mean abundances in each group
        cd_1 = X_abundances_log[cd_mask, idx1].mean()
        cd_2 = X_abundances_log[cd_mask, idx2].mean()
        control_1 = X_abundances_log[control_mask, idx1].mean()
        control_2 = X_abundances_log[control_mask, idx2].mean()
        
        # Check for opposite trends
        diff_1 = cd_1 - control_1
        diff_2 = cd_2 - control_2
        
        # If one goes down and other goes up (compensation)
        if (diff_1 * diff_2) < -0.1:  # Opposite directions
            compensation_patterns.append({
                'met_1': pair['metabolite_1'],
                'met_2': pair['metabolite_2'],
                'similarity': pair['similarity'],
                'change_1': diff_1,
                'change_2': diff_2,
                'compensation_score': abs(diff_1 - diff_2)
            })
    
    if compensation_patterns:
        comp_df = pd.DataFrame(compensation_patterns).sort_values('compensation_score', ascending=False)
        print(f"\nFound {len(comp_df)} potential compensation patterns!")
        print("\nTop 5 compensation patterns:")
        print(comp_df.head().to_string(index=False))
    else:
        print("\nNo strong compensation patterns detected.")

# ============================================================================
# 2. CHEMICAL DIVERSITY ANALYSIS
# ============================================================================
print("\n" + "#"*80)
print("# TASK 2: Chemical Diversity Profiling")
print("#"*80)

def calculate_chemical_diversity(sample_abundances, embeddings):
    """
    Calculate chemical diversity using Shannon entropy in embedding space
    High diversity = many different chemical structures
    """
    # Weight embeddings by abundance
    weights = sample_abundances / (sample_abundances.sum() + 1e-10)
    
    # Project to lower dimensions
    pca_temp = PCA(n_components=10)
    emb_reduced = pca_temp.fit_transform(embeddings)
    
    # Calculate pairwise distances (weighted by abundance)
    weighted_distances = []
    for i in range(len(embeddings)):
        for j in range(i+1, len(embeddings)):
            dist = np.linalg.norm(emb_reduced[i] - emb_reduced[j])
            weight = weights[i] * weights[j]
            weighted_distances.append(dist * weight)
    
    # Diversity score (mean weighted distance)
    diversity = np.mean(weighted_distances) if weighted_distances else 0
    return diversity

# Calculate diversity for each sample
print("\nCalculating chemical diversity for each sample...")
diversities = []

for i in range(X_abundances.shape[0]):
    sample_abund = X_abundances[i]
    diversity = calculate_chemical_diversity(sample_abund, metabolite_embeddings_array)
    diversities.append(diversity)

diversities = np.array(diversities)

# Compare diversity across groups
diversity_results = pd.DataFrame({
    'Sample': metabolite_data_matched.columns,
    'Diversity': diversities,
    'Diagnosis': labels['diagnosis_multi']
})

print("\nChemical Diversity by Diagnosis:")
for diagnosis in ['CD', 'UC', 'Control']:
    mask = diversity_results['Diagnosis'] == diagnosis
    mean_div = diversity_results[mask]['Diversity'].mean()
    std_div = diversity_results[mask]['Diversity'].std()
    print(f"  {diagnosis:10s}: {mean_div:.4f} ± {std_div:.4f}")

# Statistical test
from scipy.stats import f_oneway, kruskal
stat, pval = kruskal(
    diversity_results[diversity_results['Diagnosis'] == 'CD']['Diversity'],
    diversity_results[diversity_results['Diagnosis'] == 'UC']['Diversity'],
    diversity_results[diversity_results['Diagnosis'] == 'Control']['Diversity']
)
print(f"\nKruskal-Wallis test: H={stat:.3f}, p={pval:.4f}")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Boxplot
ax = axes[0]
diversity_results.boxplot(column='Diversity', by='Diagnosis', ax=ax)
ax.set_xlabel('Diagnosis', fontsize=12)
ax.set_ylabel('Chemical Diversity Score', fontsize=12)
ax.set_title('Chemical Diversity by Diagnosis', fontsize=13, fontweight='bold')
plt.sca(ax)
plt.xticks(rotation=0)

# Plot 2: Scatter with age
ax = axes[1]
age_valid = ~np.isnan(labels['age'])
colors = {'CD': 'red', 'UC': 'blue', 'Control': 'green'}
for diagnosis in ['CD', 'UC', 'Control']:
    mask = (diversity_results['Diagnosis'] == diagnosis) & age_valid
    if mask.sum() > 0:
        ax.scatter(labels['age'][mask], diversities[mask], 
                  c=colors[diagnosis], label=diagnosis, alpha=0.6, s=60)

ax.set_xlabel('Age', fontsize=12)
ax.set_ylabel('Chemical Diversity Score', fontsize=12)
ax.set_title('Chemical Diversity vs Age', fontsize=13, fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/chemical_diversity.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Chemical diversity analysis complete!")

# ============================================================================
# 3. PATIENT STRATIFICATION BY CHEMICAL PROFILE
# ============================================================================
print("\n" + "#"*80)
print("# TASK 3: Patient Stratification by Chemical Structure Profile")
print("#"*80)

# Create "chemical profile" for each patient using embedding-weighted features
# Use the PCA-reduced weighted embeddings we created earlier
print("\nClustering patients by their chemical profiles...")

from sklearn.cluster import KMeans, AgglomerativeClustering

# Determine optimal number of clusters using silhouette score
silhouette_scores = []
K_range = range(2, 8)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_pred = kmeans.fit_predict(X_emb_pca_scaled)
    score = silhouette_score(X_emb_pca_scaled, labels_pred)
    silhouette_scores.append(score)

optimal_k = K_range[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters (by silhouette): {optimal_k}")

# Cluster using optimal k
kmeans_chem = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
chem_clusters = kmeans_chem.fit_predict(X_emb_pca_scaled)

# Analyze cluster composition
print(f"\nChemical Profile Clusters (k={optimal_k}):")
print("="*80)

cluster_analysis = []
for cluster_id in range(optimal_k):
    mask = chem_clusters == cluster_id
    n_samples = mask.sum()
    
    # Diagnosis distribution
    cd_pct = (labels['diagnosis_multi'][mask] == 'CD').sum() / n_samples * 100
    uc_pct = (labels['diagnosis_multi'][mask] == 'UC').sum() / n_samples * 100
    ctrl_pct = (labels['diagnosis_multi'][mask] == 'Control').sum() / n_samples * 100
    
    # Age distribution
    age_valid = ~np.isnan(labels['age'][mask])
    mean_age = labels['age'][mask][age_valid].mean() if age_valid.sum() > 0 else np.nan
    
    cluster_analysis.append({
        'Cluster': cluster_id,
        'N': n_samples,
        'CD_pct': cd_pct,
        'UC_pct': uc_pct,
        'Control_pct': ctrl_pct,
        'Mean_Age': mean_age
    })
    
    print(f"\nCluster {cluster_id} (n={n_samples}):")
    print(f"  CD: {cd_pct:.1f}%, UC: {uc_pct:.1f}%, Control: {ctrl_pct:.1f}%")
    if not np.isnan(mean_age):
        print(f"  Mean age: {mean_age:.1f}")

cluster_df = pd.DataFrame(cluster_analysis)

# Visualize clusters
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: PCA of chemical profiles
ax = axes[0]
pca_viz = PCA(n_components=2)
X_pca_2d = pca_viz.fit_transform(X_emb_pca_scaled)

scatter = ax.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=chem_clusters, 
                     cmap='tab10', alpha=0.6, s=60, edgecolors='black', linewidth=0.5)
ax.set_xlabel(f'PC1 ({pca_viz.explained_variance_ratio_[0]:.1%})', fontsize=12)
ax.set_ylabel(f'PC2 ({pca_viz.explained_variance_ratio_[1]:.1%})', fontsize=12)
ax.set_title('Patient Clusters by Chemical Profile', fontsize=13, fontweight='bold')
plt.colorbar(scatter, ax=ax, label='Cluster')
ax.grid(alpha=0.3)

# Plot 2: Cluster composition
ax = axes[1]
x = np.arange(optimal_k)
width = 0.25

ax.bar(x - width, cluster_df['CD_pct'], width, label='CD', color='red', alpha=0.7)
ax.bar(x, cluster_df['UC_pct'], width, label='UC', color='blue', alpha=0.7)
ax.bar(x + width, cluster_df['Control_pct'], width, label='Control', color='green', alpha=0.7)

ax.set_xlabel('Chemical Profile Cluster', fontsize=12)
ax.set_ylabel('Percentage (%)', fontsize=12)
ax.set_title('Diagnosis Distribution in Chemical Clusters', fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels([f'C{i}' for i in range(optimal_k)])
ax.legend()
ax.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/chemical_clustering.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Chemical profile clustering complete!")

# ============================================================================
# 4. STRUCTURAL SIGNATURE ANALYSIS
# ============================================================================
print("\n" + "#"*80)
print("# TASK 4: Disease-Specific Structural Signatures")
print("#"*80)

# Find which embedding dimensions are most associated with disease
print("\nIdentifying structural features associated with disease...")

# For each embedding dimension, calculate association with disease
embedding_importance = []

for dim in range(metabolite_embeddings_array.shape[1]):
    # Get this dimension for all metabolites
    emb_dim = metabolite_embeddings_array[:, dim]
    
    # Weight by abundance difference (CD vs Control)
    cd_abundances = X_abundances_log[cd_mask].mean(axis=0)
    control_abundances = X_abundances_log[control_mask].mean(axis=0)
    abundance_diff = cd_abundances - control_abundances
    
    # Correlation between embedding dimension and abundance change
    corr = np.corrcoef(emb_dim, abundance_diff)[0, 1]
    embedding_importance.append({
        'dimension': dim,
        'correlation': abs(corr)
    })

emb_importance_df = pd.DataFrame(embedding_importance).sort_values('correlation', ascending=False)

print("\nTop 10 structural dimensions associated with disease:")
print(emb_importance_df.head(10).to_string(index=False))

# Visualize top dimensions
top_dims = emb_importance_df.head(5)['dimension'].values

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for plot_idx, dim in enumerate(top_dims):
    ax = axes[plot_idx]
    
    # Get weighted embeddings for this dimension across samples
    dim_values = []
    for i in range(X_abundances.shape[0]):
        weights = X_abundances_log[i] / (X_abundances_log[i].sum() + 1e-10)
        weighted_dim = (metabolite_embeddings_array[:, dim] * weights).sum()
        dim_values.append(weighted_dim)
    
    # Box plot by diagnosis
    plot_data = pd.DataFrame({
        'Value': dim_values,
        'Diagnosis': labels['diagnosis_multi']
    })
    
    plot_data.boxplot(column='Value', by='Diagnosis', ax=ax)
    ax.set_xlabel('Diagnosis', fontsize=10)
    ax.set_ylabel(f'Weighted Embedding Dim {dim}', fontsize=10)
    ax.set_title(f'Structural Feature {dim}\n(Correlation: {emb_importance_df.iloc[plot_idx]["correlation"]:.3f})', 
                fontsize=11, fontweight='bold')
    plt.sca(ax)
    plt.xticks(rotation=0)

# Last plot: summary
ax = axes[5]
ax.bar(range(len(top_dims)), emb_importance_df.head(5)['correlation'].values, alpha=0.7)
ax.set_xlabel('Embedding Dimension Rank', fontsize=11)
ax.set_ylabel('|Correlation| with Disease', fontsize=11)
ax.set_title('Top Structural Signatures', fontsize=12, fontweight='bold')
ax.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('/home/ubuntu/ibd_data/structural_signatures.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Structural signature analysis complete!")

print("\n" + "="*80)
print("ALL EMBEDDING-SPECIFIC ANALYSES COMPLETE!")
print("="*80)
print("\nKey Insights:")
print("  1. Metabolic compensation patterns reveal structural substitutions")
print("  2. Chemical diversity differs across disease states")
print("  3. Patients cluster by chemical profile beyond just diagnosis")
print("  4. Specific structural features are disease signatures")
print("\nThese analyses are ONLY possible with embeddings, not raw abundances!")
```

```python
# ============================================================================
# ANALYSIS 2: STRUCTURAL SIMILARITY NETWORK (with proper class coloring)
# ============================================================================
def structural_similarity_network(df4, embeddings, compare_with_tsne=True):
    """
    Build a network where edges connect structurally similar metabolites.
    Nodes are colored by metabolite class using the classification function.
    """
    print("\n" + "="*80)
    print("ANALYSIS 2: STRUCTURAL SIMILARITY NETWORK")
    print("="*80)
    
    # Get embeddings
    emb_cols = [col for col in embeddings.columns if col.startswith('emb_') or 
                (isinstance(col, (int, float)) and col not in ['Metabolite', 'SMILES'])]
    if len(emb_cols) == 0:
        emb_cols = embeddings.select_dtypes(include=[np.number]).columns.tolist()
    
    emb_matrix = embeddings[emb_cols].values
    emb_metabolites = embeddings.index.values
    
    print(f"Embedding dimensions: {emb_matrix.shape[1]}")
    print(f"Number of metabolites with embeddings: {len(emb_metabolites)}")
    
    # Standardize embeddings
    scaler = StandardScaler()
    emb_scaled = scaler.fit_transform(emb_matrix)
    
    # Calculate pairwise distances
    print("Calculating pairwise structural distances...")
    distances = squareform(pdist(emb_scaled, metric='euclidean'))
    
    # Match to df4
    metabolite_pattern = r'^(C18|C8|HILIC)'
    is_metabolite = df4.index.str.contains(metabolite_pattern, na=False, regex=True)
    metabolite_rows = df4[is_metabolite]
    
    # Create mapping
    met_to_emb_idx = {}
    for i, emb_met in enumerate(emb_metabolites):
        for df4_met in metabolite_rows.index:
            if str(emb_met) in str(df4_met) or str(df4_met) in str(emb_met):
                met_to_emb_idx[df4_met] = i
                break
    
    print(f"Mapped {len(met_to_emb_idx)} metabolites from df4 to embeddings")
    
    # Classify metabolites - get actual metabolite names from embeddings
    metabolite_classes = {}
    for met in met_to_emb_idx.keys():
        # Get the embedding index
        emb_idx = met_to_emb_idx[met]
        
        # Get the actual metabolite name from embeddings
        if 'Metabolite' in embeddings.columns:
            # Use the Metabolite column if it exists
            metabolite_name = str(embeddings.iloc[emb_idx]['Metabolite'])
        else:
            # Use the index name
            metabolite_name = str(emb_metabolites[emb_idx])
        
        # Classify using the actual metabolite name
        metabolite_classes[met] = classify_metabolite(metabolite_name)
    
    class_counts = pd.Series(metabolite_classes).value_counts()
    print(f"\nMetabolite class distribution:")
    for cls, count in class_counts.items():
        print(f"  {cls}: {count}")
    
    # Create network: connect top 5% most similar
    threshold = np.percentile(distances[distances > 0], 5)
    print(f"\nNetwork threshold (top 5% most similar): {threshold:.2f}")
    
    G = nx.Graph()
    for met in met_to_emb_idx.keys():
        G.add_node(met, class_label=metabolite_classes[met])
    
    for met1 in met_to_emb_idx.keys():
        for met2 in met_to_emb_idx.keys():
            if met1 < met2:
                idx1, idx2 = met_to_emb_idx[met1], met_to_emb_idx[met2]
                dist = distances[idx1, idx2]
                if dist < threshold:
                    G.add_edge(met1, met2, weight=1/dist)
    
    print(f"Network: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")
    
    # Project to 2D using PCA for visualization
    pca = PCA(n_components=2)
    emb_2d_pca = pca.fit_transform(emb_scaled)
    
    # Also do t-SNE for comparison
    if compare_with_tsne:
        print("Computing t-SNE for comparison...")
        tsne = TSNE(n_components=2, random_state=42, perplexity=30)
        emb_2d_tsne = tsne.fit_transform(emb_scaled)
    
    # Get node positions
    node_positions_pca = {}
    node_positions_tsne = {}
    for met, emb_idx in met_to_emb_idx.items():
        node_positions_pca[met] = emb_2d_pca[emb_idx]
        if compare_with_tsne:
            node_positions_tsne[met] = emb_2d_tsne[emb_idx]
    
    # Color map for classes - use distinct colors
    unique_classes = sorted(set(metabolite_classes.values()))
    # Use a colormap with enough distinct colors
    if len(unique_classes) <= 10:
        colors = plt.cm.tab10(np.linspace(0, 1, len(unique_classes)))
    elif len(unique_classes) <= 20:
        colors = plt.cm.tab20(np.linspace(0, 1, len(unique_classes)))
    else:
        colors = plt.cm.Set3(np.linspace(0, 1, len(unique_classes)))
    
    class_to_color = dict(zip(unique_classes, colors))
    
    # Visualize
    if compare_with_tsne:
        fig, axes = plt.subplots(1, 2, figsize=(20, 8))
    else:
        fig, axes = plt.subplots(1, 1, figsize=(10, 8))
        axes = [axes]
    
    # Main network plot (PCA-based)
    ax = axes[0]
    
    # Draw edges first
    for edge in G.edges():
        pos1, pos2 = node_positions_pca[edge[0]], node_positions_pca[edge[1]]
        ax.plot([pos1[0], pos2[0]], [pos1[1], pos2[1]], 
               'gray', alpha=0.2, linewidth=0.5, zorder=1)
    
    # Draw nodes colored by class
    for cls in unique_classes:
        nodes_in_class = [n for n in G.nodes() if metabolite_classes[n] == cls]
        if nodes_in_class:
            positions = np.array([node_positions_pca[n] for n in nodes_in_class])
            ax.scatter(positions[:, 0], positions[:, 1], 
                      c=[class_to_color[cls]], label=cls, 
                      s=30, alpha=0.7, zorder=2, edgecolors='black', linewidths=0.5)
    
    ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)', fontsize=12)
    ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)', fontsize=12)
    ax.set_title('Structural Similarity Network (PCA projection)\n(Edges = top 5% most similar by embedding distance)', 
                fontsize=13, fontweight='bold')
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8, ncol=1)
    ax.grid(True, alpha=0.3)
    
    # t-SNE comparison
    if compare_with_tsne:
        ax2 = axes[1]
        # Draw same network but with t-SNE positions
        for edge in G.edges():
            pos1, pos2 = node_positions_tsne[edge[0]], node_positions_tsne[edge[1]]
            ax2.plot([pos1[0], pos2[0]], [pos1[1], pos2[1]], 
                    'gray', alpha=0.2, linewidth=0.5, zorder=1)
        
        for cls in unique_classes:
            nodes_in_class = [n for n in G.nodes() if metabolite_classes[n] == cls]
            if nodes_in_class:
                positions = np.array([node_positions_tsne[n] for n in nodes_in_class])
                ax2.scatter(positions[:, 0], positions[:, 1], 
                          c=[class_to_color[cls]], label=cls, 
                          s=30, alpha=0.7, zorder=2, edgecolors='black', linewidths=0.5)
        
        ax2.set_xlabel('t-SNE 1', fontsize=12)
        ax2.set_ylabel('t-SNE 2', fontsize=12)
        ax2.set_title('Same Network (t-SNE projection)\n(Note: t-SNE distances not meaningful)', 
                     fontsize=13, fontweight='bold')
        ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8, ncol=1)
        ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    return G, fig, metabolite_classes, node_positions_pca

# Execute
G, fig1, metabolite_classes, node_positions = structural_similarity_network(df4, embeddings, compare_with_tsne=True)
plt.show()
```
